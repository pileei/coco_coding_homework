```python
import matplotlib.pyplot as plt
import numpy as np 
```

Task 1 

```python
mu, sigma = 0, 0.1 
x = np.random.normal(mu, sigma, 1000)

mu0, sigma0 = 0, 0.15 
y = np.random.normal(mu0, sigma0, 1000)

mu1, sigma1 = 0, 0.30

z = np.random.normal(mu1, sigma1, 1000)
```

```python
# X variable
count, bins, ignored = plt.hist(x, 30, density=True, color = "yellow")
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
                np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
                linewidth=2, color='black', label= "sigma=0.1"),
# Y variable
count, bins, ignored = plt.hist(y, 30, density=True, color = "pink")
plt.plot(bins, 1/(sigma0 * np.sqrt(2 * np.pi)) *
                np.exp( - (bins - mu0)**2 / (2 * sigma0**2) ),
                linewidth=2, color='black', label= "sigma=0.15")
# Z variable
count, bins, ignored = plt.hist(z, 30, density=True, color = "orange")
plt.plot(bins, 1/(sigma1 * np.sqrt(2 * np.pi)) *
                np.exp( - (bins - mu1)**2 / (2 * sigma1**2) ),
                linewidth=2, color='black', label= "sigma=0.30")
         
# Agrega los nombres de los ejes y titulo
plt.xlabel('X')
plt.ylabel('g(x)')
plt.title("Gaussian function")

# Agrega la leyenda
plt.legend()

plt.show()
```

```python
from pathlib import Path
```

```python
central_park_data_file = Path("/home/jupyter-maria.gonzalez67@u-21bc3/CoCoLessons/data/temperature-central-park.dat")
                               
central_park_data_file
```

```python
datos = np.loadtxt(central_park_data_file)
datos
```

```python
#Dividir el array de datos. 

lectura_id, temperaturas = datos[:, 0], datos[:, 1]
```

```python
#Convertir la columna de temperaturas de F a C. 

T = temperaturas 
tempC = (np.round((5*T/9 - 5*32/9),2))
```

```python
plt.plot(lectura_id, tempC, "o-")
plt.show()
```

```python
plt.plot(lectura_id, tempC, "o-", label="Cambio en temperatura °C ")

# Ejes
plt.xlabel("Lectura_ID")
plt.ylabel("Temperatura °C")

# Titulo 
plt.title("Temperatura en el Parque Central de NY durante el Muestreo")

#Leyenda 
plt.legend()

plt.show()
```
